```python
from IPython.display import display, Javascript, clear_output
from google.colab.output import eval_js
from base64 import b64decode
import easyocr
import cv2
import numpy as np
import ipywidgets as widgets

# Function to capture an image using your webcam in Colab.
def take_photo(filename='photo.jpg', quality=0.8):
    js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture Photo';
      div.appendChild(capture);
      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});
      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();
      // Resize the output to match the video stream.
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);
      // Wait for the user to click the button.
      await new Promise((resolve) => capture.onclick = resolve);
      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
    data = eval_js('takePhoto({})'.format(quality))
    binary = b64decode(data.split(',')[1])
    with open(filename, 'wb') as f:
        f.write(binary)
    return filename

# Function to convert Bengali digits to English digits.
def convert_bengali_to_english(text):
    mapping = {
        '০': '0', '১': '1', '২': '2', '৩': '3', '৪': '4',
        '৫': '5', '৬': '6', '৭': '7', '৮': '8', '৯': '9'
    }
    for bn, en in mapping.items():
        text = text.replace(bn, en)
    return text

# Function to clean and evaluate the equation.
def evaluate_equation(equation):
    # Keep only allowed characters: digits, basic math operators, parentheses, and decimal points.
    allowed_chars = "0123456789+-*/(). "
    cleaned = ''.join(ch for ch in equation if ch in allowed_chars)
    try:
        result = eval(cleaned)
    except Exception as e:
        result = f"Error: {e}"
    return cleaned, result
```

```python
# ----- Manual Calculation Mode -----
# Text input for manual equation entry.
manual_text = widgets.Text(
    placeholder="Enter equation here (e.g., 2+3*(4-1) or with Bengali digits like ২+৩)",
    description="Equation:",
    layout=widgets.Layout(width='80%')
)

# Button to trigger calculation.
manual_button = widgets.Button(description="Calculate", button_style='success')

# Output area to display results.
manual_output = widgets.Output()

def manual_mode(b):
    eq = manual_text.value
    eq = convert_bengali_to_english(eq)
    cleaned, result = evaluate_equation(eq)
    manual_output.clear_output()
    with manual_output:
        print("Input equation:", eq)
        print("Cleaned equation:", cleaned)
        print("Result:", result)

manual_button.on_click(manual_mode)

# ----- Camera Calculation Mode -----
# Button to capture image and calculate.
camera_button = widgets.Button(description="Capture & Calculate", button_style='info')
camera_output = widgets.Output()

def camera_mode(b):
    camera_output.clear_output()
    with camera_output:
        print("Capturing photo. Please click 'Capture Photo' in the prompt.")
    filename = take_photo()  # Capture image from webcam.
    # Read the captured image.
    img = cv2.imread(filename)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Initialize EasyOCR to detect both English and Bengali text.
    reader = easyocr.Reader(['en', 'bn'])
    result_list = reader.readtext(gray, detail=0)
    equation = ' '.join(result_list)
    equation = convert_bengali_to_english(equation)
    cleaned, result = evaluate_equation(equation)
    camera_output.clear_output()
    with camera_output:
        print("Detected equation:", equation)
        print("Cleaned equation:", cleaned)
        print("Result:", result)

camera_button.on_click(camera_mode)

# Create a Tab widget to switch between modes.
tab = widgets.Tab(children=[
    widgets.VBox([manual_text, manual_button, manual_output]),
    widgets.VBox([camera_button, camera_output])
])
tab.set_title(0, "Manual Calculation")
tab.set_title(1, "Camera Calculation")
display(tab)
```
